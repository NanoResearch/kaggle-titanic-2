```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re

deckp = re.compile('[A-Z]\d+')

def cabin(cabin):
    if ' ' not in cabin:
        return cabin
    cabins = cabin.split(' ')
    if len(cabins) == 1:
        return cabins[0]
    elif deckp.match(cabins[0]):
        return cabins[0]
    else:
        return cabins[1]

def sexnum(sex):
    if sex == 'female':
        return 1
    return 0
    
train_df = pd.read_csv('train.csv')
train_df['Sex'] = train_df.Sex.apply(sexnum)
median_male_age = train_df.Age[train_df.Sex == 0].median()
mean_male_age = train_df.Age[train_df.Sex == 0].mean()

median_female_age = train_df.Age[train_df.Sex == 1].median()
mean_female_age = train_df.Age[train_df.Sex == 1].mean()

print median_male_age, mean_male_age
print median_female_age, mean_female_age

train_df['Cabin'] = train_df.Cabin.fillna('U0')
train_df['Cabin'] = train_df.Cabin.apply(cabin)
train_df = train_df.dropna()
train_df['Deck'] = train_df.Cabin.apply(lambda c: (ord(c[0]) - ord('A')))
#train_df['Number'] = train_df.Cabin.apply(lambda c: c.split(' ')[0][1:])
#train_df['HasEngine'] = train_df.Deck.apply(lambda d: d not in ["A", "B", "C", "D", "U"])

fig = plt.figure(figsize(18,8), dpi=1600)

ax = plt.subplot2grid((1,2),(0,0))
train_df.Deck[(train_df.Survived == 1) & (train_df.Sex == 1) & (train_df.Deck <= 7)].plot(kind='kde', xlim=(0,7), color='black')
train_df.Deck[(train_df.Survived == 0) & (train_df.Sex == 1) & (train_df.Deck <= 7)].plot(kind='kde', xlim=(0,7), color='red')
ax.set_xticklabels(['A', 'B', 'C', 'D', 'E', 'F', 'G'])
title("Female Survival by Deck")
legend(('Surived', 'Died', 'Total'), loc='best')

ax = plt.subplot2grid((1,2),(0,1))  
train_df.Deck[(train_df.Survived == 1) & (train_df.Sex == 0) & (train_df.Deck < 7)].plot(kind='kde', xlim=(0,6), color='black')
train_df.Deck[(train_df.Survived == 0) & (train_df.Sex == 0) & (train_df.Deck < 7)].plot(kind='kde', xlim=(0,6), color='red')
ax.set_xticklabels(['A', 'B', 'C', 'D', 'E', 'F', 'G'])
title("Male Survival by Deck")
legend(('Surived', 'Died'), loc='best')
```

```python
from sklearn import neighbors
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.cross_validation import cross_val_score

def sex_bucket(sex):
    if sex == 'female':
        return 1
    else:
        return 0

df = train_df
median_age = train_df.Age.median()
df['Age'] = df.Age.fillna(median_age)
df['Number'] = df.Number.fillna(0)
df['SexBucket'] = df.Sex.apply(sex_bucket)
```

```python
knc = neighbors.KNeighborsClassifier(weights='distance')
feature_cols = ['Pclass', 'Age', 'SexBucket']
print cross_val_score(knc, df[feature_cols].values, df['Survived'].values).mean()

et = ExtraTreesClassifier(n_estimators=100, max_depth=None, min_samples_split=1, random_state=0)
print cross_val_score(et, df[feature_cols].values, df['Survived'].values).mean()
```
